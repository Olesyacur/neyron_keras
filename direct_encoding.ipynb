```python
import numpy as np
'''Прямое кодирование на уровне слов (упрощенно).'''
# Исходные данные: один элемент - одн образец (в данном случае это предложения, но могут быть и документы)
samples = ['The cat sat on the mat.', 'The dog eat my homework.']

token_index = {} # Создание индекса всех токенов в данных
for sample in samples:
    # Токенизация образцов с помощью метода split. Также желательно было бы удалить все знаки препинания и спец символы
    for word in sample.split():
        if word not in token_index:
            # Присваивание уникального индекса каждому уникальному слову (индекс=0 не используется)
            token_index[word] = len(token_index) + 1

# Векторзация образцов. В каждом образце рассматриваются только первые max_lenth слов
max_lenth = 10

results = np.zeros( # Сохраняем результаты
    shape = (
        len(samples),
        max_lenth,
        max(token_index.values()) + 1)
)

for i, samples in enumerate(samples):
    for j, word in list(enumerate(samples.split()))[:max_lenth]:
        index = token_index.get(word)
        results[i, j, index] = 1.

print(results)

'''Прямое кодирование на уровне символов (упрощенно).'''
import string

samples = ['The cat sat on the mat.', 'The dog eat my homework.']
#Все отображаемые символы ASCII
characters = string.printable


token_index = dict(zip(characters, range(1, len(characters)+1)))

max_length = 50
results = np.zeros(
    (len(samples),
     max_length,
     max(token_index.values())+1)
)
for i,sample in enumerate(samples):
    for j, character in enumerate(sample):
        index = token_index.get(character)
        results[i, j, index] = 1

print(results)
```
